# Klasifikasi Menggunakan Foto

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from keras.preprocessing import image
```

```python
from keras.models import load_model

model = load_model("D:/Code/Notebook/realtime-facemask-detection-cnn/model_without_opt.h5", compile = False)
```

```python
path = "D:/Code/Notebook/realtime-facemask-detection-cnn/images"

plt.style.use("default")
fig, ax = plt.subplots(nrows = 1, ncols = 6, figsize = (18, 18))

for x, img in enumerate(os.listdir(path)):
    img_path = "{}{}{}".format(path, "/", img)
    img = image.load_img(img_path, target_size = (150, 150))
    
    ax[x].imshow(img)
    
    img = image.img_to_array(img)
    img = np.expand_dims(img, axis = 0)
    
    result = model.predict(img)[0][0]
    
    if result == 0:
        ax[x].set_title("{}: With Mask".format(result))
    else:
        ax[x].set_title("{}: Without Mask".format(result))
```
